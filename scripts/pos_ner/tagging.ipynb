```python
#import relevant nltk data, functions

import nltk
from nltk import word_tokenize,pos_tag
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
# Read in a directory of txt files as the corpus using the os library.

import os
corpusdir = '/Users/bcritt/Documents/StanfordProjects/Corpora/Emerson/emerson/'
corpus = []
for infile in os.listdir(corpusdir):
    with open(corpusdir+infile, errors='ignore') as fin:
        corpus.append(fin.read())
```

```python
# Do parts of speech tagging of corpus
pos = pos_tag(word_tokenize(corpus[0]))
pos
```

```python
# Do named entity tagging of POS text
ne = nltk.ne_chunk(pos)
```

```python
# see data
print(ne)
```

# Spacy Tagging
Spacy provides tagging and is a bit more interchangeable than NLTK

```python
#spacy implementation of NER
!pip3 install spacy && python3 -m spacy download en_core_web_sm
import spacy 
nlp = spacy.load("en_core_web_sm")
```

```python
# tag text with POS info
doc = nlp(corpus[0])
pos_output = []
for token in doc:
    pos_output.append(token.text + " " + token.tag_ + " " + token.pos_)
```

```python
pos_output
```

## Convert POS output to dataframe

```python
pos_df = pd.DataFrame(pos_output)
```

```python
pos_df[['word','pos_tag', 'pos']] = pos_df[0].str.split(" ", 3, expand=True)
pos_df = pos_df.drop(pos_df.columns[0], axis=1)
```

```python
pos_df
```

```python
# tag corpus with NER data
doc = nlp(corpus[0])
ner_output = []
for token in doc:
    ner_output.append(token.text + " " + token.ent_type_)
```

```python
ner_output
```

## Convert NER output to dataframe and split columns to allow for easier cleaning

```python
ner_df = pd.DataFrame(ner_output)
```

```python
ner_df[['word','entity_type']] = ner_df[0].str.split(" ", 2, expand=True)
ner_df = ner_df.drop(ner_df.columns[0], axis=1)
```

```python
print(ner_df.to_string())
```

# Data Cleaning

Only preserve those rows where an NER classification was given

```python
import numpy as np
ner_df['entity_type'].replace('', np.nan, inplace=True)
```

```python
ner_df.dropna(subset=['entity_type'], inplace=True)
```
